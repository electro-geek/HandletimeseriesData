```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla = pdr.get_data_yahoo('TSLA')
```

```python
df_tesla.head()
```

```python
#let's plot the time series data
df_tesla.plot()
```

```python
df_tesla['High'].plot(figsize=(12,4))
```

```python
##x limit and y limit 
df_tesla['High'].plot(xlim=['2022-01-01', '2022-10-10'], ylim=[0,900], figsize=(12,4))
```

```python
##x limit and y limit and coloring
df_tesla['High'].plot(xlim=['2020-01-01', '2022-10-10'], ylim=[0,500], figsize=(12,4), c= 'green')
```
